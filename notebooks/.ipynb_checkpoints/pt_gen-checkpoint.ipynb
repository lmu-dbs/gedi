# Generating traces using pt_gen
Comparison with log from [running-example.xes](https://pm4py.fit.fraunhofer.de/static/assets/examples/running-example.xes)

```python
import pm4py
#INPUT_PATH = "~/Downloads/12716609/review_example_large.xes"
INPUT_PATH = "../data/running-example.xes"

log = pm4py.read_xes(INPUT_PATH)
```

```python
from pm4py.algo.discovery.inductive import algorithm as inductive_miner
from pm4py.algo.discovery.heuristics import algorithm as heuristics_miner
from pm4py.visualization.petri_net import visualizer as pn_visualizer
from pm4py import save_vis_petri_net

#OUTPUT_PATH ="../plots/"+INPUT_PATH.rsplit("/",1)[1].rsplit(".",1)[0]+"_all.png"
net, initial_marking, final_marking = inductive_miner.apply(log,
                                                  parameters={heuristics_miner.Variants.CLASSIC.value.Parameters.DEPENDENCY_THRESH: 1,
                                                              pn_visualizer.Variants.FREQUENCY.value.Parameters.FORMAT: "png"})
parameters = {pn_visualizer.Variants.FREQUENCY.value.Parameters.FORMAT: "png"}
fig_pt = pn_visualizer.apply(net, initial_marking, final_marking, parameters=parameters, variant=pn_visualizer.Variants.FREQUENCY, log=log)
#fig_pt = pn_visualizer.apply(net, initial_marking, final_marking, parameters=parameters, log=log)


pn_visualizer.view(fig_pt)
#print("Saved in", OUTPUT_PATH)
#save_vis_petri_net(net, initial_marking, final_marking, OUTPUT_PATH)
```

```python
runex_tree =pm4py.discover_process_tree_inductive(log)
#pm4py.view_process_tree(runex_tree)
```

```python
# Event log stats
import numpy as np
import statistics
from pm4py.objects.conversion.log import converter as log_converter

def get_length_trace_descriptors(log):
    trace_len = []
    
    variants = pm4py.get_variants_as_tuples(log)
    variant_keys = variants.keys() 
    for variant_key in variant_keys:
        #nOf_events += len(variant_key)
        for i in range(len(variants[variant_key])):
            trace_len.append(len(variants[variant_key][0]))
    min_act = min(trace_len)
    max_act = max(trace_len)
    mode = statistics.mode(trace_len)
    mean = round(np.mean(trace_len))
    std = round(np.std(trace_len),2)
    
    return min_act, max_act, mode, std

def get_nOf_events(log):
    nOf_events=0
    for trace in log: 
        #print(trace)
        nOf_events += len(trace)

    return nOf_events

def print_stats(log):
    df = log_converter.apply(log, variant=log_converter.Variants.TO_DATA_FRAME)
    print("Number of events:", get_nOf_events(log))
    print("Number of traces:", len(log))
    print("Number of variants:", len(pm4py.stats.get_variants_as_tuples(log).keys()))
    print("Number of activities:", len(df['concept:name'].unique()))
    print("Length of traces descriptors (min, max, mode, std):", get_length_trace_descriptors(log))
```

```python
playout_log = pm4py.sim.play_out(runex_tree)

net, initial_marking, final_marking = inductive_miner.apply(playout_log,
                                                  parameters={heuristics_miner.Variants.CLASSIC.value.Parameters.DEPENDENCY_THRESH: 1,
                                                              pn_visualizer.Variants.FREQUENCY.value.Parameters.FORMAT: "png"})
parameters = {pn_visualizer.Variants.FREQUENCY.value.Parameters.FORMAT: "png"}
fig_pt = pn_visualizer.apply(net, initial_marking, final_marking, parameters=parameters, variant=pn_visualizer.Variants.FREQUENCY, log=playout_log)
#fig_pt = pn_visualizer.apply(net, initial_marking, final_marking, parameters=parameters, log=log)


pn_visualizer.view(fig_pt)
```

```python
from pm4py.algo.simulation.tree_generator import algorithm as tree_generator

PARAMS = {tree_generator.Variants.PTANDLOGGENERATOR.value.Parameters.MIN: 5,
         tree_generator.Variants.PTANDLOGGENERATOR.value.Parameters.MAX: 30,
         tree_generator.Variants.PTANDLOGGENERATOR.value.Parameters.MODE: 10}
process_tree = pm4py.generate_process_tree(parameters=PARAMS)
#pm4py.view_process_tree(process_tree)
ptgen_log = pm4py.sim.play_out(process_tree)

net, initial_marking, final_marking = inductive_miner.apply(ptgen_log,
                                                  parameters={heuristics_miner.Variants.CLASSIC.value.Parameters.DEPENDENCY_THRESH: 1,
                                                              pn_visualizer.Variants.FREQUENCY.value.Parameters.FORMAT: "png"})
parameters = {pn_visualizer.Variants.FREQUENCY.value.Parameters.FORMAT: "png"}
fig_pt = pn_visualizer.apply(net, initial_marking, final_marking, parameters=parameters, variant=pn_visualizer.Variants.FREQUENCY, log=ptgen_log)
#fig_pt = pn_visualizer.apply(net, initial_marking, final_marking, parameters=parameters, log=log)


pn_visualizer.view(fig_pt)
```
