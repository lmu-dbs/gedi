```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(columns=["log","ratio_top_20_variants", "normalized_sequence_entropy_linear_forgetting"])    
```

```python
k=0
for i in np.arange(0.2, 1.1,0.2):
    for j in np.arange(0,0.55,0.1):
        k+=1
        new_entry = pd.Series({'log':f"objective_{k}", "ratio_top_20_variants":round(i,1),
                   "normalized_sequence_entropy_linear_forgetting":round(j,1)})
        df = pd.concat([
                df, 
                pd.DataFrame([new_entry], columns=new_entry.index)]
           ).reset_index(drop=True)
    
```

```python
df.to_csv("../data/grid_objectives.csv" ,index=False)
```
